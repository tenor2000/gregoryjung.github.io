We'll focus on bank stocks and see how they progressed throughout the [financial crisis](https://en.wikipedia.org/wiki/Financial_crisis_of_2007%E2%80%9308) all the way to early 2016.

## Get the Data

First we import the libraries that will be needed.


```python
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
%matplotlib inline
```

## Data

We get stock information from these major US banks:
* Bank of America
* CitiGroup
* Goldman Sachs
* JPMorgan Chase
* Morgan Stanley
* Wells Fargo

**The stock data is from Jan 1st 2006 to Jan 1st 2016 for each of these banks. Set each bank to be a separate dataframe, with the variable name for that bank being its ticker symbol.**

We will use the pickle data file provided.

```python
df = pd.read_pickle('all_banks')
```

```python
#Bank of America
BAC = df['BAC']
#CitiGroup
C = df['C']
#Goldman Sachs
GS = df['GS']
#JPMorgan Chase
JPM = df['JPM']
#Morgan Stanley
MS = df['MS']
#Wells Fargo
WFC = df['WFC']
```

```python
df.head()
```

**Create a list of the ticker symbols (as strings) in alphabetical order.**

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

**Use pd.concat to concatenate the bank dataframes together to a single data frame. Set the keys argument equal to the tickers list.**

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
```

```python
bank_stocks.head()
```

**Set the column name levels.**

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

# EDA

Let's explore the data a bit!

**What is the max Close price for each bank's stock throughout the time period?**

```python
for tick in tickers:
    print(tick,bank_stocks[tick]['Close'].max())
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

**Create a new empty DataFrame called 'returns'. This dataframe will contain the returns for each bank's stock. Returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

**Use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

**Fill 'returns' with % changes to represent the return value in each column.**

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
returns.head()
```

**Let's use a piarplot to visualize the data in order to spot any anomolies.**

```python
# Cut index 0 because of the null value in the first row 2006-01-03.
sns.pairplot(returns[1:])
```

**Let's take a look at the standard deviation of the returns and see which banks were the riskiest overall and also in 2015, the last full year of data on hand.**

```python
# It appears that Citigroup is the riskiest overall
returns.std()
```

```python
# In 2015, both Bank of American and Morgan Stanley had the highest risk, with MS slightly higher.
returns['2015-01-01':'2015-12-31'].std()
```

**Let's look at Morgan Stanley for the year 2015**

```python
sns.histplot(returns.loc['2015-01-01':'2015-12-31']['MS Return'], color='green',bins= 50,kde='True')
```

**Let's look at Citigroup in 2008 during the financial crisis**

```python
sns.histplot(returns['2008-01-01':'2008-12-31']['C Return'],color='red', bins=50, kde='True')
```

# More Visualization of the Data

An exploratory look at financial data

### Imports

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

#import plotly
#import cufflinks as cf
#cf.go_offline()
```

**Using a line plot, we'll see the Closing price for each bank for the entire index of time. This will give us an overall cross section of the data.**

```python
for tick in tickers:
    bank_stocks[tick]['Close'].plot(label=tick,figsize=(12,4))
plt.legend()
```

```python
# Alternative way using .xs
bank_stocks.xs(key='Close',axis=1,level='Stock Info').plot()
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

**Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
plt.figure(figsize=(12,6))
BAC['Close'].loc['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label='30 Day Mov Avg')
BAC['Close'].loc['2008-01-01':'2009-01-01'].plot(label='BAC Close')
plt.legend()
```

**Let's find if there are any correlations between stocks closing prices.**

```python
sns.heatmap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True)
```

**Use seaborne to cluster the information together to highlight some of the correlations**

```python
sns.clustermap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True)
```
